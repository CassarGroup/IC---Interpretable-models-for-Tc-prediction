The article "Predicting superconducting transition temperature through advanced machine learning and innovative feature engineering" (Gashmard, Shakeripour and Alaei, 2024) proposes two tools: One, for the creation of 322 features based on the chemistry formulas of the materials on the original dataset (Jabir) and other for reducing the dimensionality of the dataset obtained as Jabir ouptut (Soraya). So, in this notebook, the materials present at the dataset "Superconductivity Data" (unique_m.csv) of UCI Machine Learning Repository will be used as output for Jabir. After that, the 80 most important features will be used for training the model Clustering + GLM. The result will be compared to the same model trained with the original data (train.csv).
</p>
